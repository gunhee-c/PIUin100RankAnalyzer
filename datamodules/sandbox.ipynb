```python
import pandas as pd

songlist_origin = pd.read_csv('songList_tofill.csv')

songlist_origin
```

```python
songlist_temp = pd.read_csv('songList_temp.csv')

korean_songnames = []
korean_songnames_proxy = []
for i in range(len(songlist_temp["song"])):
    for j in range(len(songlist_origin["song"])):
        if songlist_temp["song"][i] == songlist_origin["song"][j]:
            korean_songnames.append(songlist_origin["kr_song"][j])
            korean_songnames_proxy.append(songlist_origin["kr_song_proxy"][j])
            break

print(len(songlist_temp["song"]))
print(len(korean_songnames))

songlist_temp["kr_song"] = korean_songnames
songlist_temp["kr_song_proxy"] = korean_songnames_proxy

songlist_temp.to_csv('songList_temp.csv', index=False)
```

```python
def name_in_songlist(songlist, name, index):
    if name.lower() in songlist["song"][index].lower():
        return True
    if name.lower() in songlist["kr_song"][index].lower():
        return True
    if name.lower() in songlist["kr_song_proxy"][index].lower():
        return True
    return False

def song_finder(songlist, name):
    anslist = []
    for i in range(len(songlist["song"])):
        if name_in_songlist(songlist, name, i):
            anslist.append(songlist["song"][i])
    return anslist
```
